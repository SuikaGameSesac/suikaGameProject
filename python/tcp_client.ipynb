```python
import socket
import time
host = "127.0.0.1"
port = 65432

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as socket_connect:
    socket_connect.connect((host, port))
    print("서버에 연결되었습니다.")

    while True:
        data = socket_connect.recv(65536)  # 서버로부터 데이터를 지속적으로 받음
        if not data:
            print("서버로부터 데이터 수신이 끊겼습니다.")
            break  # 데이터 수신이 없으면 연결 종료
        print(f"서버로부터 받은 데이터: {data[:20]}")
        print(f"서버로부터 받은 데이터: {len(data)}")

        # 타임 슬립 추가 (예: 1초마다 데이터 전송)
        time.sleep(0.1)
```

```python
import cv2
import numpy as np

# 디코드된 이미지를 표시하는 함수
def display_image(data):
    # 바이너리 데이터를 numpy 배열로 변환
    nparr = np.frombuffer(data, np.uint8)
    
    # numpy 배열을 이미지로 디코드
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    
    # 이미지 표시
    cv2.imshow("Received Image", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 예시 코드에서 받은 데이터를 이미지로 표시
display_image(data)
```
